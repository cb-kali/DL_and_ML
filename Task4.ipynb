```python
sns.heatmap(df.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

```python
sns.catplot(data = df,x ='Age',y='PassengerId',kind="bar",height = 6,aspect =5)
plt.xticks(rotation = 90)
plt.show()
```
